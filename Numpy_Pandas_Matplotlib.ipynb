```python
# create new col in df
# ddition of SL,SW,PL,PW
df['Total']=df['sepal_length']+df['sepal_width']+df['petal_length']+df['petal_width']
df.head()
```

```python
# Student task
# Create 6 col in given DF
# SL+SW = sepal_length + sepal_width
df['SL+SW'] = df['sepal_length'] + df['sepal_width']
df.head()
```

```python
# PL+PW = petal_length + petal_width
df['PL+PW'] = df['petal_length'] + df['petal_width']
df.head()
```

```python
# SL*PW = sepal_length * petal_width
df['SL*PW'] = df['sepal_length'] * df['petal_width']
df.head()
```

```python
# min = min([sepal_length ,sepal_width	,petal_length	,petal_width])
df['min']=df[['sepal_length','sepal_width','petal_length','petal_width']].min(axis=1)
df.tail()
```

```python
df['max']=df[['sepal_length','sepal_width','petal_length','petal_width']].max(axis=1)
df.tail()
```

```python
# mean = mean([sepal_length ,sepal_width	,petal_length	,petal_width])
df['mean']=df[['sepal_length','sepal_width','petal_length','petal_width']].mean(axis=1)
df.head()
```

```python
# Delete column in df
# axis = 1 (col)
# axis = 0 (row)
# inplace = True (perment delete)
df.drop('Total',axis = 1,inplace = True)
df.head(2)
```

```python
#Student task
# Write a code to delete all the newly created columns in df
df.drop(['SL+SW','PL+PW','SL*PW','min','max','mean'],axis=1,inplace=True)
df.head(2)
```
